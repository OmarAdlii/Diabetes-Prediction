```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, log_loss, jaccard_score, confusion_matrix
```

```python
dir=os.path.join('diabetes.csv')
data=pd.read_csv(dir)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.groupby('Outcome').mean()
```

```python
X=data.drop(columns='Outcome',axis=1)
y=data['Outcome'].copy()
```

```python
X
```

```python
y
```

```python
X=StandardScaler().fit(X).transform(X)
X
```

```python
X.shape
```

```python
type(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
lr_classifier=LogisticRegression().fit(x_train,y_train)
print(lr_classifier)
svm_classifier=SVC(kernel='linear').fit(x_train,y_train)
print(svm_classifier)
knn_classifier=KNeighborsClassifier(n_neighbors=7).fit(x_train,y_train)
print(knn_classifier)
tree_classifier=DecisionTreeClassifier(criterion='entropy',max_depth=10).fit(x_train,y_train)
print(tree_classifier)
nb_classifier=GaussianNB().fit(x_train,y_train)
print(nb_classifier)
```

```python
print(f'Logistic Regression accuracy = {accuracy_score(lr_classifier.predict(x_train),y_train)}')
print(f'Naive Bayes accuracy = {accuracy_score(nb_classifier.predict(x_train),y_train)}')
print(f'KNN accuracy = {accuracy_score(knn_classifier.predict(x_train),y_train)}')
print(f'SVM accuracy = {accuracy_score(svm_classifier.predict(x_train),y_train)}')
print(f'Decision Tree accuracy = {accuracy_score(tree_classifier.predict(x_train),y_train)}')
```

```python
lr_predictions=lr_classifier.predict(x_test)
knn_predictions=knn_classifier.predict(x_test)
svm_predictions=svm_classifier.predict(x_test)
nb_predictions=nb_classifier.predict(x_test)
tree_predictions=tree_classifier.predict(x_test)
```

```python
methods=[]
results=[]
metrices=['Accuracy','F1_Score','Jaccard Score','Log Loss']
```

```python
lr_accuracy=accuracy_score(y_test,lr_predictions)
lr_f1_score=f1_score(y_test,lr_predictions)
lr_jaccard=jaccard_score(y_test,lr_predictions)
lr_log_loss=log_loss(y_test,lr_predictions)
methods.append('Logistic Regression')
results.append([lr_accuracy,lr_f1_score,lr_jaccard,lr_log_loss])
print('Logistic Regression results')
print('accuracy score = %.2f'%lr_accuracy)
print('f1 score = %.2f'%lr_f1_score)
print('jaccard score = %.2f'%lr_jaccard)
print('log loss = %.2f'%lr_log_loss)
```

```python
knn_accuracy=accuracy_score(y_test,knn_predictions)
knn_f1_score=f1_score(y_test,knn_predictions)
knn_jaccard=jaccard_score(y_test,knn_predictions)
knn_log_loss=log_loss(y_test,knn_predictions)
methods.append('KNN')
results.append([knn_accuracy,knn_f1_score,knn_jaccard,knn_log_loss])
print('KNN results')
print('accuracy score = %.2f'%knn_accuracy)
print('f1 score = %.2f'%knn_f1_score)
print('jaccard score = %.2f'%knn_jaccard)
print('log loss = %.2f'%knn_log_loss)
```

```python
svm_accuracy=accuracy_score(y_test,svm_predictions)
svm_f1_score=f1_score(y_test,svm_predictions)
svm_jaccard=jaccard_score(y_test,svm_predictions)
svm_log_loss=log_loss(y_test,svm_predictions)
methods.append('Support Vector Machine')
results.append([svm_accuracy,svm_f1_score,svm_jaccard,svm_log_loss])
print('SVM results')
print('accuracy score = %.2f'%svm_accuracy)
print('f1 score = %.2f'%svm_f1_score)
print('jaccard score = %.2f'%svm_jaccard)
print('log loss = %.2f'%svm_log_loss)
```

```python
nb_accuracy=accuracy_score(y_test,nb_predictions)
nb_f1_score=f1_score(y_test,nb_predictions)
nb_jaccard=jaccard_score(y_test,nb_predictions)
nb_log_loss=log_loss(y_test,nb_predictions)
methods.append('Naive Bayes')
results.append([nb_accuracy,nb_f1_score,nb_jaccard,nb_log_loss])
print('Naive Bayes results')
print('accuracy score = %.2f'%nb_accuracy)
print('f1 score = %.2f'%nb_f1_score)
print('jaccard score = %.2f'%nb_jaccard)
print('log loss = %.2f'%nb_log_loss)
```

```python
tree_accuracy=accuracy_score(y_test,tree_predictions)
tree_f1_score=f1_score(y_test,tree_predictions)
tree_jaccard=jaccard_score(y_test,tree_predictions)
tree_log_loss=log_loss(y_test,tree_predictions)
methods.append('Decision Tree')
results.append([tree_accuracy,tree_f1_score,tree_jaccard,tree_log_loss])
print('Decision Tree results')
print('accuracy score = %.2f'%tree_accuracy)
print('f1 score = %.2f'%tree_f1_score)
print('jaccard score = %.2f'%tree_jaccard)
print('log loss = %.2f'%tree_log_loss)
```

```python
def plot_confusion_matrix(y,y_predict,model_name):
    cm=confusion_matrix(y,y_predict)
    ax=plt.subplot()
    sns.heatmap(cm,annot=True,ax=ax)
    ax.set_xlabel('Pridected_labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix ---->('+model_name+')')
    ax.xaxis.set_ticklabels(['Non-Diabetic','Diabetic'])
    ax.yaxis.set_ticklabels(['Non-Diabetic','Diabetic'])
    plt.show()
```

```python
plot_confusion_matrix(y_test,lr_predictions,'Logistic Regression')
```

```python
plot_confusion_matrix(y_test,svm_predictions,'Support vector machine')
```

```python
plot_confusion_matrix(y_test,knn_predictions,'KNN')
```

```python
plot_confusion_matrix(y_test,nb_predictions,'Naive Bayes')
```

```python
plot_confusion_matrix(y_test,tree_predictions,'Decision Tree')
```

```python
def comparision_methods(results,matrix,methods):
    fig=plt.figure(figsize=(8,4))
    print(results)
    plt.bar(methods,results,color=['b','g','r','c','m'])
    plt.xlabel("Methods")
    plt.ylabel(matrix)
    plt.title("Comparison of methods")
    plt.show()
```

```python
comparision_methods([row[0] for row in results],metrices[0],methods)
```
